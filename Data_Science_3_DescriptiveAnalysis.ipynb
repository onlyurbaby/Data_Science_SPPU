```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
# Drop rows with missing Age or Fare values
df_clean = df[['Sex', 'Age', 'Fare']].dropna()
```

```python
# Group by 'Sex' and get summary stats
grouped_stats = df_clean.groupby('Sex').agg({
    'Age': ['mean', 'median', 'min', 'max', 'std'],
    'Fare': ['mean', 'median', 'min', 'max', 'std']
})
```

```python
print("Grouped Summary Statistics by Sex:")
print(grouped_stats)
```

```python
# Create a list of Fare values for each Sex group
fare_by_sex = df_clean.groupby('Sex')['Fare'].apply(list).to_dict()
print("\nFare List Grouped by Sex:")
print(fare_by_sex)
```

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# Load iris dataset
iris = load_iris(as_frame=True)
df_iris = iris.frame
df_iris['species'] = df_iris['target'].map(dict(zip(range(3), iris.target_names)))
```

```python
# Group by species and calculate statistics
species_stats = df_iris.groupby('species').describe(percentiles=[.25, .5, .75])
print("Detailed Statistics Grouped by Species:\n")
print(species_stats)
```

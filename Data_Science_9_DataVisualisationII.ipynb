```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
# Box plot: age distribution by sex and survival
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
# From this plot, we can observe the following:

# 1. General Age Distribution:
# Both males and females have a wide age range, but most passengers are in their 20s to 40s.

# Median age is slightly higher for males than females in both survived and non-survived groups.

# 2. Survival Insight by Age & Gender:
# Females who survived tend to be younger, with most survivors under 40.

# Males who did not survive show a slightly broader range of ages, but survivors were generally younger.

# Infants and children seem more likely to have survived (especially among females).

# 3. Interquartile Range (IQR):
# The box (middle 50%) is tighter for females who survived, indicating they belonged to a more consistent younger age group.

# Males show a slightly wider IQR, particularly in the non-survivor group.
```
